COMPUTING PROBABILITIES

```python
# Probability of males and females buying the products.
np.round(pd.crosstab(df['Product'],[df['Gender']], normalize=True, margins=True, margins_name='Total'),2)
```

Marginal Probabilities:
P(KP281) = 0.44 i.e there is 44% chance of customers buying KP281 treadmill.
P(KP481) = 0.33 i.e there is 33% chance of customers buying KP481 treadmill.
P(KP781) = 0.22 i.e there is 22% chance of customers buying KP781 treadmill.
P(Female) = 0.42 i.e there is 42% chance that a customer buying the treadmills is a female.
P(Male) = 0.58 i.e there is 58% chance that a customer buying the treadmills is a male.

Conditional Probabilities:
P(KP281|Male) = 0.22 i.e there is 22% chance that a customer buying KP281 treadmill is a male.
P(KP481|Male) = 0.17 i.e there is 17% chance that a customer buying KP481 treadmill is a male.
P(KP781|Male) = 0.18 i.e there is 18% chance that a customer buying KP781 treadmill is a male.

P(KP281|Female) = 0.22 i.e there is 22% chance that a customer buying KP281 treadmill is a female.
P(KP481|Female) = 0.16 i.e there is 16% chance that a customer buying KP481 treadmill is a female.
P(KP781|Female) = 0.04 i.e there is 4% chance that a customer buying KP781 treadmill is a female.

```python
# Probability of customers who are partnered or single buying the products.
np.round(pd.crosstab(df['Product'],[df['MaritalStatus']], normalize=True, margins=True, margins_name='Total'),2)
```

Marginal Probabilities:
P(Partnered) = 0.59 i.e there is 59% chance that a customer buying the treadmills could be partnered.
P(Single) = 0.41 i.e there is 41% chance that a customer buying the treadmills could be single.

Conditional Probabilities:
P(KP281|Partnered) = 0.27 i.e there is 27% chance that a customer buying KP281 treadmill is partnered.
P(KP481|Partnered) = 0.20 i.e there is 20% chance that a customer buying KP481 treadmill is partnered.
P(KP781|Partnered) = 0.13 i.e there is 13% chance that a customer buying KP781 treadmill is partnered.

P(KP281|Single) = 0.18 i.e there is 18% chance that a customer buying KP281 treadmill is single.
P(KP481|Single) = 0.13 i.e there is 13% chance that a customer buying KP481 treadmill is a single.
P(KP781|Single) = 0.09 i.e there is 9% chance that a customer buying KP781 treadmill is a single.

```python
# Probability of using the products throughout the week by customers who are partnered or single.
np.round(pd.crosstab(df['MaritalStatus'],[df['Usage']], normalize=True, margins=True, margins_name='Total'),2)
```

Insights:
The above table gives us marginal and conditional probabilities of partnered and single customers and their usage of the treadmills over the days of the week in the range of 2 to 7.

Maximum marginal probabilities:
P(3) = 0.38 i.e the probability of a customer will use the treadmill 3 days a week is 38%.
P(4) = 0.29 i.e the probability of a customer will use the treadmill 4 days a week is 29%.

Maximum conditional probabilities:
P(3|Partnered) = 0.22 i.e i.e the probability of a customer who is partnered will use the treadmill 3 days a week is 22%.
P(3|Single) = 0.16 i.e the probability of a customer who is single will use the treadmill 3 days a week is 16%.

```python
# Probability of partnered/single customers rating themselves considering their fitness on a scale of 1 to 5 
np.round(pd.crosstab(df['Gender'],[df['Fitness']], normalize=True, margins=True, margins_name='Total'),2)
```

Insight:
Above table gives us marginal and conditional probabilities of customers and their self rated fitness in the range of 1 to 5.

Maximum Marginal Probabilities:
P(3) = 0.54 i.e the probability that a customer will rate their fitness 3 is 54%.
P(5) = 0.17 i.e the probability that a customer will rate their fitness 4 is 17%.

Maximum Conditional Probabilities:

BUSINESS RECOMMENDATIONS

Focus on capturing a larger customer base between the ages of 18 to 35, as this age group constitutes over 85% of treadmill buyers.

The male demographic shows a higher inclination towards treadmill purchases compared to females. Both genders show an equal likelihood of purchasing KP281 and KP481 models, but KP781 is preferred more by males, accounting for 82% of its total sales. So more attractive treadmill designs and features can be added to capture attention of both genders.

Aerofit should conduct market research to assess the potential of attracting customers with incomes under $40K, thus expanding its customer base. Basic treadmills can be created for such customers to get them hooked, so they can become potential customers for other costlier treadmills.

Position KP781 as a premium model and target marketing efforts towards higher-income groups and individuals with over 18 years of education to boost sales.

KP281 and KP481 should be promoted more for customers with median income ranges of about $60K by showcasing its affordability and showcasing it as a long term investment.

From observations we can conclude that a huge customer base is partnered, so marketing strategies should be made accordingly to promote couple based advantages for the treadmills, and for singles promotion can be done based on latest social trends.

Aerofit should promote the importance of a healthy lifestyle by showing the advantages of their treadmills for younger and older poeple and attract customers.
